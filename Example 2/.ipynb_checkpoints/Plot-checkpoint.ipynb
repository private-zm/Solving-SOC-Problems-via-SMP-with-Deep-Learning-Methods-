```python

Results3 = pd.read_csv('./LQ_data_d20.csv', index_col = 0, header = None)
Algo3 = pd.concat([Results3[i] for i in range(3,13)], axis = 0 , ignore_index= True)
Algo3.plot()
plt.title("Evolution du cost")
plt.show()
```

```python
import numpy as np

# Assuming Algo3 is a (1000, N) array where N is the number of columns
Crazy = np.zeros_like(Algo3[:100])  # Create a new array with shape (100, N)

for i in range(100):  # Loop over the new 100-row table
    for j in range(10):  # Sum over 10 segments
        Crazy[i] += Algo3[j * 100 + i]  # Add corresponding values
    Crazy[i] /= 10  # Take the mean over 10 segments
```
